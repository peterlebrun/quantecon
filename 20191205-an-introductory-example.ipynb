```python
def generate_data(n):
    _values = []
    for i in range(n):
        _values.append(np.random.rand())
    return _values

data = generate_data(100)
plt.plot(data)
plt.show()
```

```python
def generate_data(n, generator_type):
    _values = []
    for i in range(n):
        if generator_type == 'U':
            e = np.random.uniform(0, 1)
        else:
            e = np.random.randn()
        _values.append(e)
    return _values

data = generate_data(100, 'U')
plt.plot(data)
plt.show()
```

```python
# Exercise 1
def factorial(n):
    product = 1
    for i in range(n):
        product = product * (i+1)
    return product

factorial(1)
factorial(2)
factorial(3)
factorial(4)
```

```python
# Exercise 2
from numpy.random import uniform

def binomial_rv(n, p):
    draws = []
    for i in range(n):
        draws.append((uniform() < p) * 1)
    return sum(draws)/n

for i in range(10):
    print(binomial_rv(10000, 0.3))
```

```python
# Exercise 3
import numpy as np
radius = 0.5

def distance_from_center(x, y):
    return np.sqrt((0.5-x)**2 + (0.5-y)**2)

def approximate(n):
    U = np.random.uniform
    draws = []
    for i in range(n):
        draws.append(1 * (distance_from_center(U(), U()) < 0.5))
    area = sum(draws)/n
    print(area / radius**2)
                   
approximate(10**1)
approximate(10**2)
approximate(10**3)
approximate(10**4)
approximate(10**5)
approximate(10**6)
approximate(10**7)
```
